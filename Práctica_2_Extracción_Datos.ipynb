```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturación 2.xlsx')
#df
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
filtro1=df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1
```

```python
filtro2=df[~(df["CVE_VEND"] == 4.0) | (df["CVE_VEND"] == 5.0)]
filtro2
```

```python
filtro3=df[df["FECHA_ENT"] == "2022-02-28 00:00:00"]
filtro3
```

```python
filtro4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
filtro5 = df[["CAN_TOT", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]
filtro5
```

```python
filtro6= df.iloc[7001:7100,: ]
filtro6
```

```python
filtro7=df[~(df["CVE_VEND"] == 1.0) | (df["CVE_VEND"] == 2.0)]
filtro7 = df[["CVE_VEND", "FECHAELAB"]]
filtro7 = filtro7.set_index("CVE_VEND")
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Entregable2_1.csv")
filtro2.to_csv("Entregable2_2.csv")
filtro3.to_csv("Entregable2_3.csv")
filtro4.to_csv("Entregable2_4.csv")
filtro5.to_csv("Entregable2_5.csv")
filtro6.to_csv("Entregable2_6.csv")
filtro7.to_csv("Entregable2_7.csv")
```
